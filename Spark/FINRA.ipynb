Business Requirement: 

To create a report that shows, for each firm, 
the symbols they traded and the total dollar value they traded in that symbol for the period of 02/03/2014 to
02/27/2014. Use Pivot table to summarize the report.

Functional Specification: 

Table Name: Trades

This is a transaction data that represents trades in the symbol noted in the table.

Field Names:

TradeDate is the date on which the trade took place. Datatype here is DATE.

Firm is the Broker/Dealer who made the trade. Datatype here is STRING.

Symbol is the security that was traded. Datatype here is STRING.

Side represents whether the trade was a buy (purchase) or sell (sale) of the security. Datatype here is CHARECTER.

Quantity is the number of shares involved in the transaction. Datatype here is INTEGER.

Price is the price per share. Datatype here is INTEGER.

```
%sql
SELECT 
      FIRM,
      SYMBOL,
      SUM(QUANTITY * PRICE) AS TOTAL_DOLLAR_AMOUNT 
FROM TRADES 
GROUP BY 1, 2
ORDER BY 3 DESC
```
